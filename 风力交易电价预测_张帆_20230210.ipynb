# 项目背景

电价是所有市场参与者的重要的信号，在没有输电阻塞的时候，市场出清电价是整个发电侧市场中的唯一价格，但是有阻塞的时候，就采用节点电价进行结算，电价预测要做的就是预测有阻塞和没有阻塞时的节点电价。准确的电价预测是制定交易策略的关键因素，可以让市场主体在现货市场上获利更多。电价受许多因素影响，包括**市场供需、星期天、节假日、季节性、温度、突发事件、政府政策等**，导致市场主体很难通过人力判断市场电价的准确性，进而难以做出合理的现货交易策略。因此使用机器学习算法对未来电价进行预测，提高市场主体在现货市场上获利具有重要的意义。

[1.电力交易中出清电价是如何形成的?](https://www.zhihu.com/question/417051651) <br/>
在经济学中，市场出清是指当价格合适，使得市场上的任一方都能在这个价格水平卖出它拥有的东西或者买到它要的东西，并且此时市场上的需求等于供给。这个合适的价格就是指出清电价，出清电价的形成与发电侧的供给和用户侧的需求均有关

[2.节点电价和分区电价有本质的区别吗?](https://www.zhihu.com/question/385435172/answer/1130950395?utm_id=0) <br/>
节点电价是指电力市场中在**具体时间、具体地点**增加或者减少单位负荷所造成的成本变化。它反映了特定变电站在特定时间的电力供需关系，当节点电价高表示该时刻该变电站电力供需紧张，价格低就表示供需是富余的。

# 时间序列模型

<span style='color:green;font-weight:bold;font-size:1.5em'>常用的时间序列模型有四种：</span> <br/>
**自回归模型 AR(p)、移动平均模型 MA(q)、自回归移动平均模型 ARMA(p,q)、自回归差分移动平均模型 ARIMA(p,d,q)**, 前三种都是 ARIMA模型的特例 <br/>

<span style='color:green;font-weight:bold;font-size:1.5em'>AR模型</span>：
- X的期望值等于一个或数个滞后变量的线性组合，加常数项，加随机误差
- 如果自相关系数小于0.5，则不宜采用，否则预测结果极不准确
- 适用于宽平稳数据，其特性是序列的统计特性不随时间的平移而变化，即均值和协方差不随时间的平移而变化
- 在序列不平稳的时候，可以考虑用一阶差分或二阶差分去平稳数据
- 可以用ADF检验序列是否平稳(静态)

<span style='color:green;font-weight:bold;font-size:1.5em'>MA模型</span>：
- 移动平均法是根据时间序列数据逐次运行，每次都计算一定数目的时序平均数， 以一组平均值反映长期趋势
- 当时间序列的数值由于受周期变动和不规则变动的影响，起伏较大，不易显示出发展趋势时，可用移动平均法，消除这些因素的影响，显示出事件的发展方向与趋势(即趋势线)
- 与自回归模型不同，移动平均是以过去的残差项（也就是白噪声）来做线性组合，而AR模型是以过去的观察值来做线性组合。MA模型的出发点是通过组合残差项来观察残差的振动

<span style='color:green;font-weight:bold;font-size:1.5em'>ARMA模型</span>：
- AR和MA的一种组合，更普通的类型

<span style='color:green;font-weight:bold;font-size:1.5em'>ARIMA模型</span>：
- 在ARMA基础上，做差分
- 要求时序数据是稳定的（stationary），或者是通过差分化(differencing)后是稳定的
- p代表预测模型中采用的时序数据本身的滞后数(lags) ,也叫做AR/Auto-Regressive项
- d代表时序数据需要进行几阶差分化，才是稳定的，也叫Integrated项
- q代表预测模型中采用的预测误差的滞后数(lags)，也叫做MA/Moving Average项

## ADF-test检验时间序列是否静态

假设检验, H0假设模型是非静态，HA假设模型静态 <br/>
通过Augmented Dickey-Fuller Test得出p-value和ADF统计量 <br/>
如果p-value < significance-level 且 ADF统计量 < critical-value，则拒绝原假设，说明时间序列模型是静态的

```python
train2 = pd.read_csv('train2.csv', sep=',')
```

```python
train2.head()
```

## 自相关分析

Look for tail off pattern in either ACF or PACF. <br/>
- If tail off at ACF → AR model → Cut off at PACF will provide order p for AR(p).
- If tail off at PACF → MA model → Cut off at ACF will provide order q for MA(q).
- Tail of at both ACF and PACF → ARMA model

```python
# 某天数据有价格数据的采用前一个时间点插值的方法
plt.rc('figure', figsize=(10, 5))
test[test['date']=='2022/5/10'].loc[:,["open_price"]].plot(marker='o',markersize=5)
plt.ylabel("open_price")
```

```python
# 缺失值插值
# 某天数据存在煤炭价格不为空时，当天首尾值用当天mean填充，其余值用previous进行填充
from scipy.interpolate import interp1d
des_date = test.groupby('date').describe()
des_date1 = des_date['open_price']
des_date1['date'] = des_date.index
price_not_all0_date = ['2022/5/9', '2022/5/10']
for i in price_not_all0_date:
    time_order = test2.loc[(test2['date']==i), 'time_order']
    test2.loc[test2[test2['date']==i].index[0], 'open_price'] = np.array(des_date1.loc[des_date1['date']==i, 'mean'])
    test2.loc[test2[test2['date']==i].index[-1], 'open_price'] = np.array(des_date1.loc[des_date1['date']==i, 'mean'])
    
    x = test2.loc[(test2['date']==i) & (test2['open_price'].isnull().values==False), 'time_order']
    y1 = test2.loc[(test2['date']==i) & (test2['open_price'].isnull().values==False), 'open_price']
    f1 = interp1d(x, y1, kind='previous')
    test2.loc[(test2['date']==i), 'open_price'] = f1(time_order)
```

```python
plt.rc('figure', figsize=(10, 5))
test2[test2['date']=='2022/5/10'].loc[:,["open_price"]].plot(marker='o',markersize=5)
plt.ylabel("open_price")
```
